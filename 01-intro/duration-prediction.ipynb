```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('../data/yellow_tripdata_2022-01.parquet')
df.head()
```

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

df = df[(df.duration >= 1) & (df.duration <= 60)]

categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']

df[categorical] = df[categorical].astype(str)
```

```python
train_dicts = df[categorical + numerical].to_dict(orient='records')

dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)

target = 'duration'
y_train = df[target].values
```

```python
lr = LinearRegression()
Lasso = Lasso()
Ridge = Ridge()
models = [lr,Lasso,Ridge]

for _ in models:
    _.fit(X_train, y_train)
    y_pred = _.predict(X_train)
    mean_squared_error(y_train, y_pred, squared=False)
    print(f'Error for model {_} is {mean_squared_error(y_train, y_pred, squared=False)}')
    
    sns.distplot(y_pred, label=str(_) + ' ' + 'prediction')
    sns.distplot(y_train, label=str(_)+' ' +'actual')
    

    plt.legend()
```

```python
def read_dataframe(filename):
    if filename.endswith('.csv'):
        df = pd.read_csv(filename)

        df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
        df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    elif filename.endswith('.parquet'):
        df = pd.read_parquet(filename)

    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    df = df[(df.duration >= 1) & (df.duration <= 60)]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)
    
    return df
```

```python
df_train = read_dataframe('../data/yellow_tripdata_2022-01.parquet')
df_val = read_dataframe('../data/yellow_tripdata_2022-02.parquet')
```

```python
df_val.head()
```

```python
train_dicts = df_val[categorical + numerical].to_dict(orient='records')

dv = DictVectorizer()
X_val = dv.fit_transform(train_dicts)

target = 'duration'
y_val = df_val[target].values
```

```python
lr = LinearRegression()
Lasso = Lasso()
Ridge = Ridge()
models = [lr,Lasso,Ridge]

for _ in models:
    _.fit(X_train, y_train)
    y_pred = _.predict(X_val)
    mean_squared_error(y_val, y_pred, squared=False)
    print(f'Error for model {_} is {mean_squared_error(y_val, y_pred, squared=False)}')
```
